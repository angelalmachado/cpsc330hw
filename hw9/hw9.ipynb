#### Logistic regression

Our chosen predictive model for this project was **logistic regression**. This is a model used for binary classification problems, and it learns the weight each feature in the dataset has in the prediction and a bias value. The model then uses this information to make the decision of how to classify the new example we give it. In our project, the model will learn the weight different words have on making a tweet go viral or not. 

For our analysis, we first **split the data** into training and testing sets, assigning 60% of the data to the test set. The, we implemented the preprocessing steps described above (`CountVectorizer`) and ran the logistic regression model.

In this step, we also evaluated the model using **cross-validation** by splitting the training data into multiple folds, training the model on different subsets, and evaluating its average performance. The CV score obtained was 0.895.

#### Hyperparameter optimization

The goal of this step is to tune the configuration settings of our model (i.e., its hyperparameters) to improve its predictive performance. In our project, we optimized both `CountVectorizer` and `LogisticRegression` hyperparameters.

**Optimizing `max_features` of `CountVectorizer`**

The `max_features` hyperparameter controls the number of words that should be used when building the vocabulary of the model. We tested a range of possible values, from 10 up to 100,000. As shown in the graph below, 10,000 appears to be the optimal choice since it yields the highest CV score whilst keeping the gap between training and CV score at a minimum.

**Optimizing `C` of `LogisticRegression`**

The `C` hyperparameter controls the regularization of the logistic regression model, which in turn dictates how well our model will generalize to unseen data. After testing a range of values, our chosen `C` was 0.1. The same criteria used above were applied here: as seen in the plot below, 0.1 leads to the highest CV score and smallest gap between training and CV scores.
